```python
import numpy as np
```

```python
def init():
    board = np.zeros((1,9)).flatten()
    board[[3,4]] = 1
    board[[1,5]] = 2
    return board.reshape(3,3)

    
def draw_board(board):
    for y in range(board.shape[0]):
        for x in range(board.shape[1]):
            if board[y,x] == 0:
                print('   ', end="")
            elif board[y,x] == 1:
                print(' X ', end="")
            elif board[y,x] == 2:
                print(' O ', end="")
            if x != board.shape[1] - 1:
                print('|', end="")
        if y != board.shape[0] - 1:
                print('\n-----------')
    print('\n\n')
    
    
def MCTS(board):
    best_move = None
    
    return best_move
        
        
def move(board, player):
    # all possible actions
    actions = np.argwhere(board == 0)
    actions_idxs = range(len(actions))
    
    # only make a move if an action can be taken
    if actions.size != 0:
        
        # player 'X'
        if player == 0:
            best_move = MCTS(board)
            
            # select a random action
            action_idx = np.random.choice(actions_idxs, size=1)
            y,x = actions[action_idx][0]

            # do action on board
            board[y,x] = 1 if player == 0 else 2
            
        # player 'O'
        elif player == 1:
            # select a random action
            action_idx = np.random.choice(actions_idxs, size=1)
            y,x = actions[action_idx][0]

            # do action on board
            board[y,x] = 1 if player == 0 else 2
    
    return board


def check_winner(board):
    winner = None

    # check rows
    for row in board:
        if all(row == 1):
            winner = 0
        elif all(row == 2):
            winner = 1
            
    # check columns
    for column in board.T:
        if all(column == 1):
            winner = 0
        elif all(column == 2):
            winner = 1
    
    # check diagonals
    if all(np.diag(board) == 1):
        winner = 0
    elif all(np.diag(board) == 2):
        winner = 1
    if all(np.diag(np.flip(board, axis=0)) == 1):
        winner = 0
    elif all(np.diag(np.flip(board, axis=0)) == 2):
        winner = 1
    
    return winner
    


def simulate_match(board):
    # draw the initial state of the board
    draw_board(board)
    
    # initiate player that has to move
    player = 0
    
    winner = None
    while winner == None:
        
        prev_board = board.copy()

        # update the board with a move from the player
        board = move(board, player)
        
        # check whether there is a winner
        winner = check_winner(board)
        
        # board is the same as before
        if np.array_equal(prev_board, board):
            break
        
        # draw the updated board
        draw_board(board)
        
        # alternate player between 0 and 1
        player = (player + 1) % 2

    return winner
    
def main():
    board = init()
    winner = simulate_match(board)
    
    if winner == 0:
        print('Winner = X')
    elif winner == 1:
        print('Winner = O')
    else:
        print('Draw')
    
main()
```
